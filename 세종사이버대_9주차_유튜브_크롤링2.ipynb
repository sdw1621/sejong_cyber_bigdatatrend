#### **11-1. 판다스 패키지 호출, 구글 API 패키지 호출**

```python
from googleapiclient.discovery import build

# 여러 댓글 저장하는 변수
comments = list() 

# 생각보다 꽤 예쁘고 좋네요? 자잘하게 많이 개선된 삼성 갤럭시 워치4 언빡싱&첫인상!
# https://www.youtube.com/watch?v=TkHLpGLR_dc&ab_channel=ITSub잇섭
videoId="TkHLpGLR_dc"

# 구글 유튜브 API 키
# AIzaSyDYYyNxiwBYdDZC8WSef0Qi1xurnR7wekI
api_obj = build('youtube', 'v3', developerKey="AIzaSyDYYyNxiwBYdDZC8WSef0Qi1xurnR7wekI")

response = api_obj.commentThreads().list(part='snippet, replies', videoId=videoId, maxResults=100).execute()
print(response)

# 변수 참고
# commentThreads : 댓글을 한번에 여러개를 수집함(1번, 2번 시간차로 시작하는데, 끝은 순차적으로 들어오지 않을 수도 있음)
# snippet : 댓글, 처음 올린 댓글
# replies : 답장, 댓글에 달린 답장
# videoId : 영상의 아이디
# maxResults : 갯수? 퍼센트?
# execute : 실행
```

#### **11-2. 댓글 데이터 수집 후 저장**

```python
while response:  
  for item in response['items']: # items 중에서
    comment = item['snippet']['topLevelComment']['snippet']
    comments.append([comment['textDisplay'], comment['authorDisplayName'], comment['publishedAt'], comment['likeCount']])
    # 선택적으로 append
    if item['snippet']['totalReplyCount'] > 0:
      for reply_item in item['replies']['comments']:
        reply = reply_item['snippet']
        comments.append([reply['textDisplay'], reply['authorDisplayName'], reply['publishedAt'], reply['likeCount']])

	    # 댓글들 페이지 업데이트하며 불러옴
  if 'nextPageToken' in response:
    response = api_obj.commentThreads().list(part='snippet, replies', videoId=videoId, pageToken=response['nextPageToken'], maxResults=100).execute()
  else:
    break

# 변수 참고
# snippet : 댓글
# topLevelComment : 첫 번째 레벨의 댓글
# textDisplay : 유튜브 댓글에 보여지는 내용
# authorDisplayName : 채널명
# publishedAt : 댓글 쓴 날짜
# likeCount : 댓글에 좋아요 붙은 갯수
# totalReplyCount : 전체 답글 갯수
# replies : 답글
# nextPageToken : 다음 댓글 정보 있는지 여부
```
